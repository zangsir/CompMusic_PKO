```python
#sum up above
def preprocess(file):
    f=open(file,'r')
    t=f.read()
    tdec=codecs.decode(t,'utf-8')
    tres=tdec.replace('   ',',')
    tresn=tres.replace('\n','.')
    tresns=tresn.replace(' ','')
    newt=[]
    for c in tresns:
        newt.append(c)
    return newt

file='原板.txt'
res=preprocess(file)
```

```python
from collections import defaultdict
#text is a list, the product of preprocess
def ngrams(n,text):
    dic=defaultdict(int)
    for i in range(len(text)-n):
        tokens=text[i]
        for j in range(1,n):
            tokens+=" " + text[i+j]
        dic[tokens]+=1
    dit=sorted(dic.items(), key=lambda x: x[1],reverse=True)
    return dit
```

```python
def print_result(result):
    output=[]
    
    for k,v in result:
        if ',' not in k and '.' not in k:
            output.append([k,str(v)])
    return output
```

```python
out=print_result(result)
```

```python
y=open('ngrams-lyrics.txt','a')

for line in out[:100]:
    ttw=' '.join(line)
    y.write(ttw.encode('utf-8'))
    y.write('\n')
y.close()
```
