## doing collapsed shengqiang, only dividing by banshi to create docs

```python
#doing the dic again but collapse xipi_yuanban and erhuang_yuanban into yuanban, etc.
import codecs, re
per_label_collapse = defaultdict(dict)
inputfilename="xk-seg-cln-nop-fix.csv"

selected='西皮摇板,西皮慢板,西皮原板,西皮快板,二黄摇板,二黄慢板,二黄原板'
seldec=codecs.decode(selected,'utf-8')
seldec=seldec.split(",")

    
    
with open(inputfilename, 'rU') as inputfile:
    reader = csv.reader(inputfile)
 

    for play,banshi,lyrics in reader:
        pldec=codecs.decode(play,'utf-8')
        bsdec=codecs.decode(banshi,'utf-8')
        lydec=codecs.decode(lyrics,'utf-8')
        #we want only the 7 types, then collapse different shengqiang
        key_clean=bsdec.replace(u'\uff09','').replace(u'\uff08','')
        key_collapsed=key_clean[-2:]
        #print key_clean
        if key_clean in seldec:
        
            if not key_collapsed in per_label_collapse[pldec]:#the banshi doesn't exist yet in this play
                per_label_collapse[pldec][key_collapsed]=[lydec]
            else:
                per_label_collapse[pldec][key_collapsed].append(lydec)
```

```python
#print all keys from the the first two dict to see what it is like
i=1
for k in per_label_collapse:
    print k
    for key in per_label_collapse[k]:
        print key,len(per_label_collapse[k][key])
        for line in per_label_collapse[k][key]:
            print line + '\n'
    print '========================'
    i=i+1
    if i==7:
        break
```
